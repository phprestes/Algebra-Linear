### Exercício 01 - Subespaço

Considere, em $\mathbb{R}^3$, o plano dado pela equação $x+y-z=0$.
Os vetores $u=[1, 0, 1]^\top$ e $v=[0, 1, 1]^\top$ pertencem a tal plano (como verificar esta afirmação?). Qualquer combinação linear $w=au + bv$, com $a,b\in \mathbb{R}$, pertence ao plano, pois

$$
w = \begin{bmatrix} a \\ b \\ a+b \end{bmatrix}.
$$

Logo o plano corresponde a um subespaço de $\mathbb{R}^3$.
- Implemente uma visualização desse subespaço com os vetores $u$ e $v$ e uma combinação linear qualquer desses vetores.

```python
!pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Criar a figura 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')

# Vetores do subespaço
u = np.array([1, 0, 1])
v = np.array([0, 1, 1])

# Combinação linear (valores aleatórios para a e b)
a, b = 0.5, 1.5
w = a*u + b*v

# Criar o plano x + y - z = 0
x_range = np.linspace(-3, 3)
y_range = np.linspace(-3, 3)
X, Y = np.meshgrid(x_range, y_range)
Z = X + Y

# Plotar o plano
ax.plot_surface(X, Y, Z, alpha=0.3, color='lightblue', label='Plano x+y-z=0')

# Plotar os vetores
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='red', arrow_length_ratio=0.1, linewidth=2, label='u = [1, 0, 1]')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='blue', arrow_length_ratio=0.1, linewidth=2, label='v = [0, 1, 1]')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='green', arrow_length_ratio=0.1, linewidth=3, 
          label=f'w = {a}u + {b}v = [{w[0]}, {w[1]}, {w[2]}]')

# Verificar que os vetores pertencem ao plano
print(f"Verificação que u pertence ao plano: {u[0]} + {u[1]} - {u[2]} = {u[0] + u[1] - u[2]}")
print(f"Verificação que v pertence ao plano: {v[0]} + {v[1]} - {v[2]} = {v[0] + v[1] - v[2]}")
print(f"Verificação que w pertence ao plano: {w[0]} + {w[1]} - {w[2]} = {w[0] + w[1] - w[2]}")

# Configurações do gráfico
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Subespaço: Plano x + y - z = 0\ncom vetores u, v e combinação linear w = au + bv')

# Adicionar grade
ax.grid(True)

# Adicionar legenda
ax.legend()

# Mostrar o gráfico
plt.show()
```

### Exercício 02 - Mudança de base
Considere a base $B$ de $\mathbb{R}^3$ formada pelos vetores $(1,0,1)$, $(1,1,1)$ e $(1,1,2)$. Considere também a base $C$ formada pelos vetores $(1,0,0)$, $(0,1,0)$ e $(0,0,1)$.
- Calcule $M_B^C$ e $M_C^B$
- Generalize o cálculo da matriz de mudança de base em uma função.
- Sejam as coordenadas do vetor $v$ em relação a base $C$ dadas por $v_C = [8, 7, 1]^\top$. Calcule as coordenadas do vetor $v$ em relação a base $B$.

```python
B = np.array([[1, 0, 1],
             [1, 1, 1],
             [1, 1, 2]]).T

C = np.array([[1, 0, 0],
             [0, 1, 0],
             [0, 0, 1]]).T

B, C
```

```python
def mudancaBase(orig, dest):
    return np.linalg.inv(dest) @ orig

print("Matriz de mudança de base de B para C:\n", mudancaBase(B, C))
print("Matriz de mudança de base de C para B:\n", mudancaBase(C, B))
```

```python
mudancaBase(C, B) @ np.array([8, 7, 1]).T
```
