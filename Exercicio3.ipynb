### Exercício 01 - Subespaço
Considere os subespaços de $\mathbb{R}^3$ dados por
$$
U = \{(x,y,z) \in \mathbb{R}^3; x = 0 \} \quad \text{e}\quad
V = \{(x,y,z) \in \mathbb{R}^3; y = 0 \}.
$$
Observe que $\mathbb{R}^3 = U + V$, porém a soma não é direta.

Implemente uma visualização adequada para $U$, $V$ e  $U \cap V$.

```python
%pip install matplotlib
%pip install numpy
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')

# Subespaço V
x = np.linspace(-5, 5, 20)
z = np.linspace(-5, 5, 20)
X1, Z1 = np.meshgrid(x, z)
Y1 = np.zeros_like(X1)
ax.plot_surface(X1, Y1, Z1, alpha=0.5, color='green', label='Subespaço V (y=0)')

# Subespaço U
y = np.linspace(-5, 5, 20)
z = np.linspace(-5, 5, 20)
Y2, Z2 = np.meshgrid(y, z)
X2 = np.zeros_like(Y2)

# Interseção
z_intersec = np.linspace(-5, 5, 100)
x_intersec = np.zeros_like(z_intersec)
y_intersec = np.zeros_like(z_intersec)
ax.plot(x_intersec, y_intersec, z_intersec, 
        color='orange', linewidth=4, label='Interseção U∩V (eixo Z)')

ax.plot_surface(X2, Y2, Z2, alpha=0.5, color='red', label='Subespaço U (x=0)')

# Configurações do gráfico
ax.set_title('Subespaços U e V')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Melhorar a visualização
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

plt.tight_layout()
plt.show()
```

#### Exercício 02 - Transformação Linear
Seja a transformação linear $T:\mathbb{R}^2 \to \mathbb{R}^2$ associada a matriz
$$ A =
\begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
.$$
Considere a circunferência $C=\{(x,y)\in\mathbb{R}^2;\; x^2+y^2=1\}$. Implemente uma visualização da imagem da circunferência $C$ pela transformação linear $T$.
- Dica: Discretize um conjunto de pontos de $C$ através da parametrização $(x, y) = (cos(t), sen(t))$ e aplique a transformação nesses pontos.

```python
# Matriz da transformação linear
A = np.array([[2, 1],
              [1, 2]])

# Parametrização da circunferência
t = np.linspace(0, 2*np.pi)
x_c = np.cos(t)
y_c = np.sin(t)
circunferencia = np.vstack((x_c, y_c))  # Matriz 2xN
circunferencia
```

```python
# Aplicar a transformação linear
transformada = A @ circunferencia

plt.figure(figsize=(10, 5))

# Extrair coordenadas transformadas
x_t = transformada[0, :]
y_t = transformada[1, :]

# Plotar as curvas
plt.plot(x_c, y_c, 'b-', linewidth=2, label='Circunferência C')
plt.plot(x_t, y_t, 'r-', linewidth=2, label='Imagem T(C)')
plt.title('Circuferência C e Imagem T(C)')

# Configurações do plot
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
